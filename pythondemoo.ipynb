```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Study_Hours": [2.5, 3.0, 3.8, 4.2, 5.1, 6.0, 6.5],
    "Sleep_Hours": [7.5, 7.0, 6.2, 5.8, 5.0, 4.5, 4.0],
    "Coffee_Cups": [1, 1, 2, 2, 3, 4, 4]
}

df = pd.DataFrame(data)
df
```

BAR **PLOT**

```python
plt.bar(df["Day"], df["Coffee_Cups"])
plt.title("Coffee Intake vs Exam Stress")
plt.xlabel("Day")
plt.ylabel("Cups of Coffee")
plt.show()
```

scatter plot


```python
plt.scatter(df["Study_Hours"], df["Sleep_Hours"])
plt.title("Study Hours vs Sleep Hours")
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.show()
```

piechart

```python
time_spent = [sum(df["Study_Hours"]), sum(df["Sleep_Hours"])]
labels = ["Studying", "Sleeping"]

plt.pie(time_spent, labels=labels, autopct="%1.1f%%")
plt.title("Student Life Breakdown")
plt.show()
```

histogram

```python
plt.hist(df["Study_Hours"], bins=5)
plt.title("Distribution of Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Number of Days")
plt.show()

```
