```python
import numpy as np  #we import the libraries to get all the 'tools'
```

**Creating an array**

```python
a=np.array([1,3,4,8])
print(a)
```

FLOAT TYPE

```python
b = np.array([1.4, 3.2, 5], dtype = float)
print(b)
```

**INTEGER**

```python
c = np.array([1,3,5,7,9,11], dtype = int)
print(c)
```

 ***Initial*** ***Placeholders***

Create an array with zeros.

```python
np.zeros((3,8))
```

Create an array with ones.

```python
np.ones((3,8))
```

Create an array with even-spaced step values

```python
	np.arange(1, 30, 5)
```

Create an evenly-spaced array with a specified number of values

```python
np.linspace(0,4,8)
```

Create an entire array with a constant value.

```python
np.full((4,4),9)
```

Create an identity matrix.

```python
np.identity(3)
```

Create an identity matrix with m*n dimension.

```python
np.eye(3, k=1)
```

Create an array with random integers.

```python
np.random.randint(2,8,3)
```

***ARITHMETIC*** ***FUNCTIONS***

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

```

**addition**

```python
np.add(arr1,arr2)
```

**subtraction**

```python
np.subtract(arr1,arr2)
```

**multiply**

```python
	np.multiply(arr1, arr2)
```

**division**

```python
np.divide(arr1,arr2)
```

**Square** **root**

```python
np.sqrt(arr1)
```
